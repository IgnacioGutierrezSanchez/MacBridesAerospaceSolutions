```python
import pandas as pd
from pathlib import Path
import pyModeS as pms
from pyModeS import adsb
import base64
import tarfile
from datetime import datetime
import dask.dataframe as dd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

## Ejercicio 2

```python
# Ruta al archivo CSV
path = Path("output_by_day/data_2024-12-01.csv")

# Leer el archivo CSV con Dask
df_horas_1 = dd.read_csv(path)

# Convertir la columna Timestamps a datetime
df_horas_1["Timestamps"] = dd.to_datetime(df_horas_1["Timestamps"])

# Extraer la hora de la columna Timestamps
df_horas_1["Hora"] = df_horas_1["Timestamps"].dt.hour

df_horas_1['type_code'] = df_horas_1.map_partitions(
    lambda df_horas_1: df_horas_1['message'].apply(lambda msg: pms.common.typecode(msg)),
    meta=('message', 'int64') 
)

# Función para determinar la velocidad (Speed)
def determinar_speed(row):
    tc = row['type_code']
    msg = row['message']
    
    if 5 <= tc <= 8:  # Surface position
        v = adsb.surface_velocity(msg)
        if v is not None:
            return v[0]  # Speed en knots
    elif tc == 19:  # Airborne velocity
        velocity = adsb.velocity(msg)
        if velocity is not None:
            return velocity[0]  # Speed en knots
    return None

# Aplicar la función para añadir la columna 'Speed'
df_horas_1['Speed'] = df_horas_1.map_partitions(
    lambda df: df.apply(determinar_speed, axis=1),
    meta=('Speed', 'float64')
)

# Convertir la columna 'Capability' a string y luego a booleano en cada partición
df_horas_1['Capability'] = df_horas_1.map_partitions(
    lambda df: df['Capability'].astype(str).map({'True': True, 'False': False}),
    meta=('Capability', 'bool')
)

df_horas_1.head()
```

```python
df = df_horas_1.compute()
df
```

```python
# Ordenar por aeronave y tiempo
df = df.sort_values(by=['ICAO', 'Timestamps'])

# Lista para almacenar tiempos de espera y horas
tiempos_espera = []
horas_espera = []

# Recorrer cada aeronave individualmente
for icao, grupo in df.groupby('ICAO'):
    # Filtrar mensajes en tierra (Capability == False y Speed == 0)
    tierra = grupo[(grupo['Capability'] == False) & (grupo['Speed'] == 0)]
    
    # Filtrar mensajes en vuelo (Capability == True y Speed > 0)
    vuelo = grupo[(grupo['Capability'] == True) & (grupo['Speed'] > 0)]

    # Si no hay suficientes datos, continuar con la siguiente aeronave
    if tierra.empty or vuelo.empty:
        continue

    # Recorrer los mensajes en tierra
    for idx, fila_tierra in tierra.iterrows():
        # Buscar el primer mensaje en vuelo después del mensaje en tierra
        primer_vuelo = vuelo[vuelo['Timestamps'] > fila_tierra['Timestamps']].head(1)
        
        if not primer_vuelo.empty:
            # Calcular el tiempo de espera
            tiempo_espera = (primer_vuelo['Timestamps'].iloc[0] - fila_tierra['Timestamps']).total_seconds()
            
            # Almacenar el tiempo de espera y la hora del evento en tierra
            if tiempo_espera > 0:
                tiempos_espera.append(tiempo_espera)
                horas_espera.append(fila_tierra['Timestamps'].hour)

# Convertir a DataFrame
df_espera = pd.DataFrame({'Tiempo de espera': tiempos_espera, 'Hora': horas_espera})
```

```python
# Configurar el estilo de las gráficas
sns.set(style="whitegrid")

# **Histograma**
plt.figure(figsize=(10, 6))
sns.histplot(df_espera['Tiempo de espera'], bins=50, kde=True, color='blue')
plt.xlabel('Tiempo de espera (segundos)')
plt.ylabel('Frecuencia')
plt.title('Distribución de los tiempos de espera')
plt.show()
```

### Análisis del Histograma de Tiempos de Espera

1. **Asimetría positiva (sesgo a la derecha)**  
   - La mayoría de los datos están en valores bajos, con una larga cola hacia la derecha.  
   - Indica que algunos tiempos de espera son muy altos, pero poco frecuentes.

2. **Distribución sesgada**  
   - La media probablemente es mayor que la mediana debido a los valores extremos.  

3. **Valores atípicos (outliers)**  
   - Se observan algunos valores extremadamente altos (>10,000 s).  
   - Pueden ser eventos raros o errores en los datos.

4. **Densidad KDE**  
   - La línea azul muestra una estimación suave de la distribución, resaltando la alta concentración en valores bajos y la larga cola derecha.

### Conclusión  
La mayoría de los tiempos de espera son cortos, pero existen casos raros con tiempos muy largos. Si estos datos representan un sistema de servicio o transporte, podrían indicar congestión en ciertos casos extremos.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Supongamos que df_espera es tu DataFrame y "Tiempo de espera" es tu columna

# Calcular la media
media = df_espera['Tiempo de espera'].mean()
print(f"Media: {media}")

# Calcular la mediana
mediana = df_espera['Tiempo de espera'].median()
print(f"Mediana: {mediana}")

# Calcular los cuartiles
cuartil_1 = df_espera['Tiempo de espera'].quantile(0.25)
cuartil_3 = df_espera['Tiempo de espera'].quantile(0.75)
print(f"Cuartil 1: {cuartil_1}")
print(f"Cuartil 3: {cuartil_3}")

# Calcular el valor mínimo
minimo = df_espera['Tiempo de espera'].min()
print(f"Mínimo: {minimo}")

# Calcular el valor máximo
maximo = df_espera['Tiempo de espera'].max()
print(f"Máximo: {maximo}")

# Calcular el rango intercuartílico (IQR)
iqr = cuartil_3 - cuartil_1
print(f"IQR: {iqr}")

# Calcular los límites para los bigotes del boxplot (usando 1.5 * IQR)
limite_inferior = cuartil_1 - 1.5 * iqr
limite_superior = cuartil_3 + 1.5 * iqr
print(f"Límite inferior para bigotes: {limite_inferior}")
print(f"Límite superior para bigotes: {limite_superior}")

# Contar el número de valores por encima del tercer cuartil
valores_encima_cuartil3 = df_espera[df_espera['Tiempo de espera'] > cuartil_3]
cantidad_encima_cuartil3 = len(valores_encima_cuartil3)
print(f"Número de valores por encima del tercer cuartil: {cantidad_encima_cuartil3}")

# Configurar el estilo de las gráficas
sns.set(style="whitegrid")

# Crear el boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_espera['Tiempo de espera'], color='lightblue')

# Personalizar etiquetas y título
plt.xlabel('Tiempo de espera (segundos)')
plt.title('Boxplot de los tiempos de espera')

# Mostrar la gráfica
plt.show()
```

El boxplot muestra que la mayoría de los tiempos de espera son muy cortos, pero hay algunos casos en los que los tiempos de espera son significativamente mayores. La larga cola de valores atípicos a la derecha sugiere que la distribución de los tiempos de espera está sesgada hacia la derecha. Esto significa que la media de los tiempos de espera será mayor que la mediana.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar por hora y calcular el tiempo de espera promedio
espera_horas = df_espera.groupby('Hora')['Tiempo de espera'].mean().reset_index()

# Crear una tabla pivote para el mapa de calor
heatmap_data = espera_horas.set_index('Hora')

# Graficar mapa de calor horizontal
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".1f", cbar_kws={'label': 'Tiempo de espera promedio (segundos)'}, yticklabels=True)
plt.title('Mapa de calor de tiempo de espera por hora')
plt.gca().invert_yaxis()  # Invertir el eje Y para que la hora 0 esté arriba
plt.show()
```

### Análisis del Mapa de Calor del Tiempo de Espera por Hora  

1. **Tiempos de espera más altos en la madrugada (0-3 h)**  
   - Las horas entre **medianoche y las 3 a.m.** presentan los mayores tiempos de espera, con valores superiores a **19,000 segundos (~5 horas)**.  
   - Esto puede deberse a una menor disponibilidad de recursos o menor demanda, lo que genera tiempos prolongados.

2. **Disminución progresiva a partir de las 4 a.m.**  
   - Desde las **4 a.m.**, el tiempo de espera se reduce gradualmente, alcanzando su punto más bajo entre **8 y 10 a.m.**, con valores por debajo de **2,500 segundos (~40 minutos)**.  
   - Esto podría indicar un aumento en la disponibilidad del servicio o una menor congestión.

3. **Estabilidad en horas laborales (10 a.m. - 6 p.m.)**  
   - Durante el día, los tiempos de espera oscilan entre **1,500 y 3,000 segundos (~25-50 minutos)**, mostrando un comportamiento relativamente estable.  
   - Se observa un ligero pico en la **hora del almuerzo (12-1 p.m.)**, posiblemente debido a un aumento en la demanda.

4. **Menor tiempo de espera en la noche (8-9 p.m.)**  
   - A partir de las **7 p.m.**, los tiempos de espera disminuyen gradualmente, alcanzando su punto más bajo a las **8-9 p.m.** con solo **180 segundos (~3 minutos)**.  
   - Esto sugiere que la demanda es baja y el sistema opera con mayor eficiencia en este período.

### Conclusión  
El análisis sugiere que los tiempos de espera son más largos en la **madrugada**, disminuyen por la mañana y se estabilizan durante el día. En la **noche**, los tiempos de espera son significativamente más bajos, lo que puede indicar menor congestión o mayor eficiencia operativa en esas horas.
